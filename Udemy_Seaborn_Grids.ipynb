```python

sns.PairGrid(iris)
```

```python
sns.set_style('darkgrid')
sns.PairGrid(iris)
```

```python
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot,kde=False)
g.map_
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot,kde=False,bins=10)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```
