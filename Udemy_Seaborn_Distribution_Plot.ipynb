```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
## Dist Plot -> show the distribution on univariant set of observation .only take one input (one variable)
sns.distplot(tips['total_bill']) ## result is histogram iwht KDE
```

```python
sns.set_style("darkgrid")
sns.distplot(tips['total_bill'],kde=False) ## result is histogram
```

```python
sns.set_style("whitegrid")
sns.distplot(tips['total_bill'],kde=False)  ## result is histogram X-axes is count and Y-axes is bill .There are lots of bill between 10 to 20
```

```python
sns.set_style("darkgrid")
sns.distplot(tips['tip'],kde=False,bins=20)
```

```python
sns.set_style("darkgrid")
sns.distplot(tips['total_bill'],kde=False,bins=25)
```

```python
## Joint plot which will be used to compare between 2 variants (2 variables comparision)
sns.jointplot(x = 'total_bill',y = 'tip',data = tips) ## give the scatter plot
```

```python
## Create hexa plot using jointplot
sns.jointplot(x = 'total_bill',y = 'tip',data = tips,kind='hex') ## kind use type of plot we can use.. here is hexagon

```

```python
## Regression plot
sns.jointplot(x = 'total_bill',y = 'tip',data = tips,kind='reg')
```

```python
sns.jointplot(x = 'total_bill',y = 'tip',data = tips,kind = 'kde') ## 2 dimentional kde
```

```python
## Pair plot is plot created between 2 numeric variables across all the fields in data frame
sns.pairplot(tips)
```

```python
## Use the categorical variable inside the pair plot suing hue
sns.pairplot(tips,hue='sex')
```

```python
sns.pairplot(tips,hue='smoker')
```

```python
sns.rugplot(tips['total_bill']) ## build the each distrubiton in dash line
```

```python
sns.distplot(tips['total_bill'],kde=True) ## Kde= Kernel density estimation 
```

```python
## Create KDE plot
sns.kdeplot(tips['total_bill'])
```
