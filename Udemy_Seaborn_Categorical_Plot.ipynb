```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
import numpy as np
import pandas as pd
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.describe()
```

```python
## Bar Plot for categorical plotting. Average totalbill for sex. it means male has slightly average total_bill than female
sns.barplot(x = 'sex',y = 'total_bill' , data = tips) ## here default estimator is mean
```

```python
## plot a bar chart on standard deviation of total_bil   for sex (male or female)
sns.barplot(x = 'sex',y = 'total_bill',data = tips,estimator=np.std)
```

```python
## Count plot -> explicitly count based on categorical column : here one column as input which categorical column
sns.countplot(x = 'sex',data = tips)
```

```python
tips.describe()
```

```python
## Box Plot
sns.boxplot(tips.total_bill)
```

```python
sns.boxplot(tips.tip)
#comment : middle black line between box-> mean(averge) of tip (2.998279)
# Left box start -> 25% percetile(lower quartile) of tip (2.000000)
# right box end -> 75% percentile(upper quartile) of tip(3.562500)
#IQR(Inter quertile range) -> upper quartile - lower quartile=(3.562500-2.000000)=1.562500
# One whisker is drawn from the lower quartile down to the lowest data point (extreamly left black bar) 
# > lower quartile - 1.5 IQR=2.000000-1.5 * 1.562500=.34375 
# The other whisker is drawn from the upper quartile to the highest data point(extreamly right black bar) =
#>upper quartile+1.5 IQR=2+2.34375=4.34
```

```python
sns.boxplot(x=tips.sex,y=tips.total_bill)
```

```python
sns.boxplot(x=tips.day,y=tips.total_bill)
```

```python
sns.boxplot(x=tips.day,y=tips.total_bill,hue=tips.smoker)
```

```python
sns.violinplot(x=tips.day,y=tips.total_bill,hue=tips.sex)
```

```python
sns.violinplot(x=tips.day,y=tips.total_bill,hue=tips.sex,split=True)
```
