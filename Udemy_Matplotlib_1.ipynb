```python
import numpy as np
```

```python
## create list with name X which have 11 values between 0 to 5 which is equal sliced
X = np.linspace(0,5,11)
```

```python
Y = X ** 2
```

```python
X
```

```python
Y
```

```python
## Fucntional Method
plt.plot(X,Y)
plt.show()
```

```python
## Define X axis as X-Label and Y as Y-Label and title is Title
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.title('MainPlot')
plt.plot(X,Y)
```

```python
## Create 2 subplot from X,Y
## This subplot will be created in one with 2 columns strcture and frist column will be this plot
plt.subplot(1,2,1)  
plt.plot(X,Y,'r')
# This subplot will be created in one with 2 columns strcture and second column will be this plot
plt.subplot(1,2,2) 
plt.plot(Y,X,'y')
```

```python
## Object oriented method
### Create figure->its top level container which can contain all Axes and all properties of a plot
fig=plt.figure()
fig.add_axes([.1,.1,.8,.8])
```

```python
fig=plt.figure()
fig.add_axes([10,10,.8,.8])
```

```python
fig=plt.figure()
axes = fig.add_axes([10,10,.8,.8])
axes.plot(X,Y)
```

```python
fig=plt.figure()
axes = fig.add_axes([0.1,0.1,1.0,0.8])
axes.plot(X,Y)
axes.set_xlabel('X-Label')
axes.set_ylabel('Y-Label')
axes.set_title('Title')
```

```python
fig1 = plt.figure()
axes1 = fig1.add_axes([0.1,0.1,1.1,1.1])
axes2 = fig1.add_axes([0.3,0.7,0.4,0.4])
axes1.plot(X,Y,'r')
axes2.plot(Y,X,'y')
axes1.set_title('OuterPlot')
axes2.set_title('InnerPlot')
```
