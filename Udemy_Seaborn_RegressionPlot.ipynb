```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip')
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip',hue='sex')
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip',hue='smoker')
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip',hue='sex',markers=['o','v'])
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip',hue='sex',markers=['o','v'],scatter_kws={'s':50})
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip',col='time')
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip',col='time',row='day')
```
