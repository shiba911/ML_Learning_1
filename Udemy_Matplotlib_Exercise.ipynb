```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.arange(0,100)
```

```python
y = x*2
z = x**2
```

```python
fig = plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='blue')
ax.set_title('title')
ax.set_xlabel('x')
ax.set_ylabel('y')
```

```python
## Create figure which has 2 axes one is ax1 with axes [0,0,1,1] and x2 with [0.2,0.5,0.2,0.2]
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([.2,.5,.2,.2])
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([.2,.5,.2,.2])
ax1.plot(x,y,color='red')
ax2.plot(x,y,color='red')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
```

```python
##create 2 axes ax1 and ax2 and ax1 will be [0,0,1,1] and ax2 will be [.2,.5,.4,.4]
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([.2,.5,.4,.4])
```

```python
## Create 2 axes one ax1 between x and z and another ax2 between x and y but ax2 will be zoomed one
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([.2,.5,.4,.4])
ax1.plot(x,z,color='blue')
ax1.set_xlabel('X')
ax1.set_ylabel('Z')
ax2.plot(x,y,color='blue')
ax2.set_xlim([20,22])
ax2.set_ylim([30,50])
ax2.set_title('zoom')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y,color='blue',ls='--')
axes[1].plot(x,z,color='red',lw=5)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(12,2))
axes[0].plot(x,y,color='blue',lw=3)
axes[1].plot(x,z,color='red',ls='--')
```
