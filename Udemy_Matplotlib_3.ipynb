```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(0,10,10)
```

```python
x
```

```python
y = x**2
```

```python
y
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
##axes.plot(x,y,color='orange')
axes.plot(x,y,color='purple',linewidth=0.5)
## Another way to change the color using hexa decimal value for RGB..can serch google by RDB hex
###axes.plot(x,y,color='#FF8C00') 
```

```python
## use the function alph which will be change the transparent of line
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
##axes.plot(x,y,color='orange')
axes.plot(x,y,color='purple',lw=3,alpha=.25)
## Another way to change the color using hexa decimal value for RGB..can serch google by RDB hex
#axes.plot(x,y,color='#FF8C00') 
```

```python
## Add the line style
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
##axes.plot(x,y,color='orange')
axes.plot(x,y,color='purple',lw=3,alpha=1,linestyle='--')
```

```python
## Line should be in steps
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,color='red',alpha=1,linestyle='steps')
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,color='red',lw=10,alpha=.25,ls='--')
```

```python
## enable the marker
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,color='purple',lw=1,alpha=2,ls='-',marker='o',markersize=10)
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,color='purple',lw=2,alpha=.25,ls='-',marker='+',markersize=20,
          markeredgecolor='red') ## set the marker edge color
```

```python
## Set the marker face color
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,color='purple',lw=3,alpha=1,ls='--',marker='o',markersize=20,
          markerfacecolor='yellow',markeredgecolor='green')
```

```python
## Control over axis appearance and plot range

fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,color='purple',lw=3,alpha=1,ls='--')
## limit to x axes. lets say want to plot between 0 to 2 in x-axes
axes.set_xlim([0,2])
axes.set_ylim([0,20])
```
