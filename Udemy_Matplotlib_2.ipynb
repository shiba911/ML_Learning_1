```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
X = np.linspace(0,5,10)
```

```python
X
```

```python
Y = X ** 2
```

```python
Y
```

```python
fig,axes = plt.subplots(nrows = 1,ncols = 2)
## plt.tight_layout() ## remove the overlapping
axes[0].plot(X,Y,'r')
axes[0].set_title('FirstPlot')
axes[0].set_xlabel('FristX-Label')
axes[0].set_ylabel('FirstY-Label')
axes[1].plot(Y,X,'g')
axes[1].set_xlabel('SecondX-Label')
axes[1].set_ylabel('SecondY-Label')
axes[1].set_title('SecondPlot')
plt.tight_layout()
```

```python
fig = plt.figure(figsize=(4,4),dpi = 100) ## figszie is 3 inch width and 3 inch height and dpi -> resolution is 300
axes = fig.add_axes([0.1,0.1,.8,.8])
axes.plot(X,Y,'r')
axes.set_title('Title')
axes.set_xlabel('X')
axes.set_ylabel('Y')
```

```python
fig = plt.figure(figsize=(3,2),dpi = 100)
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.set_title('MainPlot')
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.plot(X,X ** 2,'r',label='X squared')
axes.plot(X,X ** 3,'g',label='X cubed')
axes.legend(loc=0)  ## loc = 0 is place in proper position
```
